Интернет-магазину товаров для дома «Пока все ещё тут» требуется полноценный анализ товарного ассортимента.

Датасет описывает транзакции интернет-магазина товаров для дома и быта «Пока все ещё тут».

Колонки в  *ecommerce_dataset.csv* :

- `date` — дата заказа;
- `customer_id` — идентификатор покупателя;
- `order_id` — идентификатор заказа;
- `product` — наименование товара;
- `quantity` — количество товара в заказе;
- `price` — цена товара.

## Исследовательский анализ данных<a id="step4"></a>
[*Назад к оглавлению*](#content)

Посмотрим, какие аномалии и выбросы есть в исходных данных. Рассмотрим столбцы с кол-вом товара в заказе и ценой:

```python
ecom[['quantity', 'price']].describe()
```

Посмотрим выбросы на диаграмме размаха:

```python
# диаграмма размаха
ecom.boxplot(column='quantity', figsize=(17,5))
plt.title('Диаграмма размаха')
plt.show()
```

```python
# диаграмма размаха - приблизим основные значения
ecom.boxplot(column='quantity', figsize=(17,5))
plt.title('Диаграмма размаха')
plt.ylim(0, 10)
plt.show()
```

Диаграмма размаха говорит о том, что все заказы в которых количество (одного и того же) товара было больше или равно двум можно считать выбросами, так как это достаточно редкие являния. Посмотрим на 95-й и 99-й перцентили столбца `quantity`

```python
percentiles = np.percentile(ecom['quantity'], [95, 99])
print('Перцентили:\n95: {}\n99: {}'.format(percentiles[0].round(),percentiles[1].round()))
```

Будем считать транзакцию аномальной, если количество товара в ней больше либо равно 27 единиц(то есть она попадает 99% перцентиль). Посмотрим в каких заказах было куплено аномально много товара:

```python
ecom[ecom['quantity'] >= 27].sort_values(by = 'quantity', ascending = False).head()
```

Из данных видно, что такие транзакции крайне редко встречаются, но могут сильно исказить статистику продаж. По сути это единичные оптовые продажи. Удалим эти выбросы из данных:

```python
# сохраним исходную таблицу
ecom_orig = copy.copy(ecom)
```

```python
# удаляем выбросы
ecom = ecom[ecom['quantity'] < 27]
```

```python
# диаграмма размаха
ecom.boxplot(column='price', figsize=(17,5))
plt.title('Диаграмма размаха')
#plt.ylim(0, 1000)
plt.show()
```

```python
# диаграмма размаха
ecom.boxplot(column='price', figsize=(17,5))
plt.title('Диаграмма размаха')
plt.ylim(0, 1000)
plt.show()
```

Диаграмма размаха для цены показывает, что обычно цена товара находятся в пределах 1 тыс. рублей, но самый дорогой из проданных товаров стоил более 14 тыс. Единичные продажи очень дорогих товаров также могут исказить статистику продаж. Посмотрим на 95-й и 99-й перцентили столбца `price`:

```python
percentiles = np.percentile(ecom['price'], [95, 99])
print('Перцентили:\n95: {}\n99: {}'.format(percentiles[0].round(),percentiles[1].round()))
```

Будем считать транзакцию аномальной, если цена товара в ней больше 5000 (то есть она попадает 99% перцентиль). Посмотрим в каких транзакциях цена была аномально большой:

```python
ecom[ecom['price'] > 5000].sort_values(by = 'price', ascending = False).head()
```

```python
ecom[ecom['price'] > 5000].sort_values(by = 'price', ascending = False).describe()
```

Из данных видно, что такие транзакции редко встречаются(51 штука), но могут сильно исказить статистику продаж. Удалим эти выбросы из данных:

```python
# удаляем выбросы
ecom = ecom[ecom['price'] <= 5000]
```

```python
ecom.head()
```

Добавим в данные столбец с датой и посчитаем с каждой транзакции выручку:

```python
# переименуем столбец date в datetime
ecom.rename(columns={'date': 'datetime'}, inplace=True)

# в столбец date запишем дату
ecom['date'] = pd.DatetimeIndex(ecom['datetime']).date

# посчитаем выручку
ecom['revenue'] = ecom['quantity'] * ecom['price']
ecom.head(10)
```

```python
ecom.describe()
```

Удалим дубликаты заказов, в которых стоят различные даты:

```python
# посмотрим на дубликаты
ecom[ecom[['customer_id', 'order_id', 'product']].duplicated(keep = False)].head()
```

```python
#удалим дубликаты
ecom = ecom[~ecom[['customer_id', 'order_id', 'product']].duplicated()]
```

```python
#проверим группировкой
ecom.groupby(['customer_id', 'order_id', 'product']).count().sort_values('date', ascending=False).head()
```

Построим сводную таблицу по датам со следующими столбцами:
 - Кол-во уникальных покупателей за день;
 - Кол-во заказов за день;
 - Средняя цена товара проданного за день;
 - Кол-во транзакций за день;
 - Кол-во товара проданного за день;
 - Суммарная выручка за день.

```python
# сводная таблица
ecom_date = ecom.pivot_table(index=['date'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_date.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 'Кол-во транзакций', 'Кол-во товара', 'Выручка']

ecom_date.head()
```

```python
ecom_date.describe()
```

Изучим количество заказов(покупок) по дням. Из таблицы видно, что в среднем соверщается примерно 7 заказов в день (медиана равна 6).

```python
ecom_date['Кол-во заказов'].describe()
```

Посмотрим на графиках, как менялось количество заказов по дням:

```python
# временной график
ecom_date['Кол-во заказов'].plot(figsize=(17,5))
plt.title('График количества заказов по дням')
plt.xlabel('Дата')
plt.ylabel('Кол-во заказов')
plt.show()
```

Хорошо заметено увеличение количества заказов перед новым годом и в начале марта.

```python
# диаграмма размаха
ecom_date.boxplot(column=['Кол-во заказов', 'Кол-во покупателей', 'Кол-во транзакций'], figsize=(17,5))
plt.title('Диаграмма размаха')
plt.ylim(0, 40)
plt.show()
```

```python
# плотность распределения
plt.figure(figsize=(17, 5))
sns.distplot(ecom_date['Кол-во заказов'], hist = False)
plt.title('График плотности распределения заказов по дням')
plt.xlabel('Кол-во заказов')
plt.ylabel('Плотность распределения')
plt.show()
```

Обычно совершается  от 5 до 7 заказов в день. Если за день совершается более 15 заказов - такой день можно считать выбросом и аномалией.

Изучим количество уникальных покупателей по дням. Из сводной таблицы видно, что в среднем каждый день примерно 7 покупателей делают заказ в магазине. В среднем на одного покупателя приходится один заказ, хотя бывают дни, когда один и тот же покупатель за день делает несколько заказов.

```python
ecom_date['Кол-во покупателей'].describe()
```

Посмотрим на графиках, как менялось количество покупателей по дням:

```python
# временной график
ecom_date['Кол-во покупателей'].plot(figsize=(17,5))
plt.title('График количества покупателей по дням')
plt.xlabel('Дата')
plt.ylabel('Кол-во покупателей')
plt.show()
```

График количества покупателей очень похож на график заказов, потому что чаще всего за день один покупатель делает только один заказ. Видна сильная корреляция между этими показателями.

```python
# диаграмма размаха
ecom_date.boxplot(column=['Кол-во покупателей', 'Кол-во заказов'], figsize=(17,5))
plt.title('Диаграмма размаха')
plt.show()
```

```python
# плотность распределения
plt.figure(figsize=(17, 5))
sns.distplot(ecom_date['Кол-во покупателей'], hist = False)
plt.title('График плотности распределения покупателей по дням')
plt.xlabel('Кол-во покупателей')
plt.ylabel('Плотность распределения')
plt.show()
```

На графиках видно, что в среднем в магазине делают покупки 4-8 уникальных покупателей в день. Если за день было больше 14 покупателей, то такой день можно назвать аномальным и посчитать выбросом.

Построим сводную таблицу по месяцам:

```python
# в столбец 'year' год
ecom['year'] = pd.DatetimeIndex(ecom['datetime']).year

# в столбец 'month' запишем месяц
ecom['month'] = pd.DatetimeIndex(ecom['datetime']).month

ecom.head()
```

```python
# сводная таблица
ecom_month = ecom.pivot_table(index=['year','month'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_month.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 
                      'Кол-во транзакций', 'Кол-во товара', 'Выручка']

ecom_month
```

```python
ecom_month.describe()
```

Проанализируем ежемесячную выручку:

```python
ecom_month['Выручка'].describe()
```

```python
ecom_month['Выручка'].sum()
```

В среднем ежемесячная выручка равна примерно 210 тыс. Самая большая выручка была в октябре 2018 года - более 320 тыс., а самая маленькая в июне 2019 - 145 тыс. рублей.

Построим график ежемесячной выручки:

```python
# временной график
ecom_month['Выручка'].plot(figsize=(17,5))
plt.title('График ежемесячной выручки')
plt.xlabel('Дата')
plt.ylabel('Рубли')
plt.show()
```

На графике видно, что пик продаж был в самом начале, в октябре 2018 года. Далее ежемесячная выручка снижалась с несколькими резкими спадами и последующими востановлениями, колеблясь в районе 200 -250 тыс. до июня 2019 года, после чего плавно снизилась до примерно 170 тыс. рублей в месяц. К сожалений, в датасете мало данных, чтобы сопоставить выручку от года к году. Но можно сравнить результаты в октябре 2018 с результатами в в октябре 2019.

```python
# выводим октябрь 2018 и октябрь 2019
year_dif = ecom_month.iloc[[0,12]].T
year_dif.columns = ['2018-10', '2019-10']
```

```python
# считаем разницу
year_dif['Разница, %'] = (year_dif['2019-10'] - year_dif['2018-10']) / year_dif['2018-10'] * 100
```

```python
year_dif
```

Из таблицы видно, что по всем основным показателям, кроме средней цены товара, у интернет-магазина идёт снижение. И если количество покупателей снизилось на 9 %, то количество заказов снизилось на 19%, количество транзакций снизилось более чем в два раза (-57%). Количество продаваемого товара также снизилось в двое (-58%). Получается, что хотя количество заказов и не так сильно снизилось, но очень сильно снизилось количество товара в одном заказе (практически в двое). Единственное, что выросло за год - это средняя цена товара (на 24 % с 554 рублей до 690). Поэтому действительно необходимо провести анализ товарного ассортимента, чтобы правильно настроить рекомендательную систему. Нет ничего удивительного, что выручка магазина за год сократилась на 48% (с 320 тыс. до 167 тыс. руб.). Замечу, что я анализирую очищенные данные, в них нет очень больших и редких покупок, но зато лучше заметна динамика продаж.

## Анализ торгового ассортимента<a id="step5"></a>
[*Назад к оглавлению*](#content)

Исследуем выручку по каждому товару за год. Убёрем из данных октября 2018 года, чтобы рассмотреть период с ноября 2018 года по октябрь 2019 (полный год).

```python
ecom_year = ecom[ecom['datetime'] >= '2018-11-01 00:00:00']
```

```python
ecom_year.head()
```

Сгруппируем товар по названию и построим свобную таблицу:

```python
# сводная таблица
ecom_prod = ecom_year.pivot_table(index=['product'], values=['revenue', 'price', 'quantity', 'datetime'], 
                    aggfunc={'revenue': 'sum', 
                             'price': 'mean', 
                             'quantity': 'sum',
                             'datetime': 'count'}).round().sort_values(by = 'revenue', ascending = False)

ecom_prod.columns = ['Кол-во заказов', 'Средняя цена', 'Кол-во товара', 'Выручка']
```

```python
ecom_prod.head()
```

```python
ecom_prod.describe()
```

```python
# диаграмма размаха
ecom_prod.boxplot(column='Выручка', figsize=(17,5))
plt.title('Диаграмма размаха')
plt.ylabel('Рубли')
#plt.ylim(0, 5000)
plt.show()
```

```python
# диаграмма размаха
ecom_prod.boxplot(column='Выручка', figsize=(17,5))
plt.title('Диаграмма размаха')
plt.ylabel('Рубли')
plt.ylim(0, 4000)
plt.show()
```

```python
# плотность распределения
plt.figure(figsize=(17, 5))
sns.distplot(ecom_prod['Выручка'], hist = False)
plt.xlim(0, 10000)
plt.title('График плотности распределения выручки для товаров')
plt.xlabel('Выручка, руб.')
plt.ylabel('Плотность распределения')
plt.show()
```

За год было продано более 2 тыс. различных товаров. Среднее значение для выручки равно 1200 рублей, а медиана примерно 450, то есть среднестатистический товар за год приносит компании 450 рублей выручки. Хотя максимальная выручка для товара за год составляет более 46 тыс. руб. (и это после очистки данных от выбросов). Лидером продаж является `Сумка-тележка 2-х колесная Gimi Argo синяя` и этот заслуженная победа, так как этот товар присутствовал в 43 различных заказах (то есть не является единоразовой оптовой продажей).

```python
ecom_prod.head()
```

```python
ecom_prod.tail()
```

Но в списке есть товары, которые принесли за год всего 9 рублей выручки, и это конечно очень мало. Встаёт вопрос о целесообразности продаж таких товаров. Также в списке не видно товаров, которых вообще никто за год не купил (хотя наверное у магазина такие товары тоже есть).

Разделим товары на две категории: основной ассортимент и дополнительный. Будем разделять товары по годовой выручке - товары, которые выше определённого значения выручки - попадают в категорию основных, а товары, которые принесли меньше определённого порога - в категорию дополнительных. Такое разделение позволит сконцентрировать внимание на товарах, которые хорошо продаются в магазине, чтобы они всегда были в наличии и на видном месте на сайте. А также подумать над сменой стратегии по продвижению дополнительного товара(с низкими продажами) или заменой его на другие товары.

Получается, что для товаров примерно выполняется Закон Парето: «20 % усилий дают 80 % результата, а остальные 80 % усилий — лишь 20 % результата». В нашем случае закон звучит примерно так: 25% от товара обеспечивают 76% выручки, а остальные 75 % товаров — лишь 24 % выручки. Такое товар имеет полное право называться основным. Разделим товары в категориях но основные и дополнительные по выбронному критерию - среднегодовой выручке.

```python
categories = ['растения', 'хозтовары', 'интерьер', 'сумки', 'другое']

for col in categories:
    ecom_col = ecom_prod.query('category == @col')
    rev_mean =  ecom_col.describe().iloc[1,3]
    print('В категории', col, 'среднее значение годовой выручки равно', round(rev_mean))
    # присваиваем значение через lambda функцию
    ecom_col['label'] = ecom_col['Выручка'].apply(lambda x : 'first' if (x > rev_mean) else 'second')
    ecom_prod.loc[ecom_col.query('label == "first"').index, 'label'] = 'first'
    ecom_prod.loc[ecom_col.query('label == "second"').index, 'label'] = 'second'
    print()
```

Посмотрим на динамику ежемесячной выручки основной группы и дополнительной для различных категорий товаров.

### Растения

```python
# код можно автоматизировать для всех категорий, если не нужны комментарии
cat = 'растения'
ecom_cat = ecom_prod.query('category ==@cat')
 
# список основных товаров
list_first = ecom_cat[ecom_cat['label'] == 'first']['product'].to_list()
list_second = ecom_cat[ecom_cat['label'] == 'second']['product'].to_list()
# сводная таблица для основного товара
ecom_month_first = ecom_year.query('product == @list_first').pivot_table(
        index=['year','month'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_month_first.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 
                      'Кол-во транзакций', 'Кол-во товара', 'Выручка']
# сводная таблица для дополнительного товара
ecom_month_second = ecom_year.query('product == @list_second').pivot_table(
        index=['year','month'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_month_second.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 
                      'Кол-во транзакций', 'Кол-во товара', 'Выручка']
# временной график
ecom_month_first['Выручка'].plot(figsize=(17,5))
ecom_month_second['Выручка'].plot(figsize=(17,5))

plt.title('График ежемесячной выручки за год для основного и дополнительного ассортимента для категории "растения"')
plt.xlabel('Дата')
plt.ylabel('Рубли')
plt.show()

# Выделим среднии показатели обеих групп:
print('Таблица сравнения показателей между основной группой и дополнительной для категории', cat)
df1_2 = pd.concat(
        [ecom_month_first.describe(),ecom_month_second.describe()]).iloc[[1,9]]
df1_2.index = ['first','second']
df1_2 = df1_2.T
df1_2['first/second'] = df1_2['first'] / df1_2['second']
df1_2.round(2)
```

Для растений наблюдается ярко выраженный пик продаж в мае. Также в мае наблюдается наиболее выраженная разница между продажами основного и дополнительного ассортимента. Из таблицы видно, что в среднем кол-во основного товара проданного за месяц в категории `растения` в 1.6 раза превышает кол-во дополнительного товара. Ежемесячная выручка больше в 2.5 раза.

```python

cat = 'другое'
ecom_cat = ecom_prod.query('category ==@cat')
 
# список основных товаров
list_first = ecom_cat[ecom_cat['label'] == 'first']['product'].to_list()
list_second = ecom_cat[ecom_cat['label'] == 'second']['product'].to_list()
# сводная таблица для основного товара
ecom_month_first = ecom_year.query('product == @list_first').pivot_table(
        index=['year','month'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_month_first.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 
                      'Кол-во транзакций', 'Кол-во товара', 'Выручка']
# сводная таблица для дополнительного товара
ecom_month_second = ecom_year.query('product == @list_second').pivot_table(
        index=['year','month'], values=['customer_id', 'order_id', 'product', 'quantity', 'price', 'revenue'], 
                    aggfunc={'customer_id': 'nunique', 
                             'order_id': 'nunique', 
                             'product': 'count',
                             'quantity': 'sum',
                             'price': 'mean',
                             'revenue': 'sum'}).round()

ecom_month_second.columns = ['Кол-во покупателей', 'Кол-во заказов', 'Средняя цена', 
                      'Кол-во транзакций', 'Кол-во товара', 'Выручка']
# временной график
ecom_month_first['Выручка'].plot(figsize=(17,5))
ecom_month_second['Выручка'].plot(figsize=(17,5))

plt.title('График ежемесячной выручки за год для основного и дополнительного ассортимента для категории "другое"')
plt.xlabel('Дата')
plt.ylabel('Рубли')
plt.show()

# Выделим среднии показатели обеих групп:
print('Таблица сравнения показателей между основной группой и дополнительной для категории', cat)
df1_2 = pd.concat(
        [ecom_month_first.describe(),ecom_month_second.describe()]).iloc[[1,9]]
df1_2.index = ['first','second']
df1_2 = df1_2.T
df1_2['first/second'] = df1_2['first'] / df1_2['second']
df1_2.round(2)
```

В категорию `другое` попали товары, которые не получилось лекго отнести к одной из выбранных категорий. Однако и для этой категории наблюдается похожая картина. Хотя по количеству продаж дополнительный ассортимент в этой категории показывает себя лучше основного, но ежемесячная выручка для основного ассортимента в более чем в три раза выше, чем ежемесячная выручка для дополнительного ассортимента.

## Проверка статистических гипотез<a id="step6"></a>
[*Назад к оглавлению*](#content)

Проверем, являются ли статистически значимыми различия между основной группой товаров и дополнительной:

```python
# стат. характеристики основной группы за год
ecom_prod[ecom_prod['label'] == 'first'].describe()
```

```python
# стат. характеристики дополнительной группы за год
ecom_prod[ecom_prod['label'] == 'second'].describe()
```

Выделим стреднии показатели обеих групп:

```python
df1_2 = pd.concat(
    [ecom_prod[ecom_prod['label'] == 'first'].describe(),ecom_prod[ecom_prod['label'] == 'second'].describe()]).iloc[[1,9]]
df1_2.index = ['first','second']
df1_2
```

```python
df1_2 = df1_2.T
df1_2['first/second'] = df1_2['first'] / df1_2['second']
df1_2
```

Проверим гипотезу, что средняя цена товара из основной группы статистически отличается от средней цены товара из дополнительной группы.

Проверим, являются ли различия в средних значениях метрик между основной группой и дополнительной группой статистически значимыми . Проверка будет проходить по следующим метрикам:
 - кол-во заказов товара за год;
 - средняя цена товара; 
 - кол-во купленного товара за год; 
 - годовая выручка для товара. 
 
Проверим следующую гипотезу: средние значения метрик отличаются между основной группой и дополнительной группой. Сформилируем нулевую гипотезу: H₀ — среднее значение метрики товаров основной группы равно среднему значению метрики товаров дополнительной группы. Альтернативная гипотеза: H₁ — среднее значение метрики товаров основной группы отличается от среднего значения метрики товаров дополнительной группы. Уровень значимости α = 5%. Необходима проверка гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. 

Перед начало проверки выберем правильный метод тестирования гипотез. Проверим наборы данных на нормальность критерием Шапиро-Уилка.

```python
#набор метрик
ecom_columns = ['Кол-во заказов', 'Средняя цена', 'Кол-во товара', 'Выручка']

for col in ecom_columns:
    print('Метрика:', col)
    sample_first = ecom_prod[ecom_prod['label'] == 'first'][col]
    sample_second = ecom_prod[ecom_prod['label'] == 'second'][col]
    
    # проверим распределения на нормальность
    alpha = .05 # критический уровень статистической значимости

    # для проверки на нормальность используем критерий Шапиро-Уилка
    results = stats.shapiro(sample_first)
    p_value = results[1] 

    print('p-значение: ', p_value)

    if (p_value < alpha):
        print("Отвергаем нулевую гипотезу: распределение метрики для основных товаров  не нормально\n")
    else:
        print("Не получилось отвергнуть нулевую гипотезу: распределение метрики для основных товаров нормально\n")

    results = stats.shapiro(sample_second)
    p_value = results[1] 

    print('p-значение: ', p_value)

    if (p_value < alpha):
        print("Отвергаем нулевую гипотезу: распределение метрики для дополнительных товаров не нормально\n")
    else:
        print("Не получилось отвергнуть нулевую гипотезу: распределение метрики для дополнительных товаров нормально\n")
        
    print('############################')
```

Для проверки гипотез воспользуемся непараметрическим тестом Манна-Уитни, так как критерий Шапиро-Уилка показал, что выборки не являются нормальными. Определим уровень значимости, который необходимо использовать для корректной проверки статистических гипотез. Необходимо будет провести 4 проверки, так как у нас 4 метрики. Таким образом, руководствуясь методом Бонферрони, легко посчитать, что уровень значимости должен быть равен 0.05/4=0.0125 (если считать, что в случае одиночного теста был бы использован уровень значимости 0.05).

```python
#набор метрик
ecom_columns = ['Кол-во заказов', 'Средняя цена', 'Кол-во товара', 'Выручка']

for col in ecom_columns:
    print('Метрика:', col)
    sample_first = ecom_prod[ecom_prod['label'] == 'first'][col]
    sample_second = ecom_prod[ecom_prod['label'] == 'second'][col]
    
    results = stats.mannwhitneyu(sample_first, sample_second)
    
    alpha = .0125 # критический уровень статистической значимости
    p_value = results[1] 

    print('p-значение: {0:.4f}'.format(p_value))

    if (p_value < alpha):
        print('Отвергаем нулевую гипотезу:', col, 'между основными и дополнительными товарами статистически различны')
    else:
        print('Не получилось отвергнуть нулевую гипотезу:', col, 
              'между основными и дополнительными товарами статистически неразличимы')
    print()
    print('#####################')
    print()
    
```

### Выводы

Непараметрический тестом Манна-Уитни показал, что различия между метриками для основной группы товаров и дополнительной являются статистически значимыми. По всем метрикам (кол-во заказов товара за год, средняя цена товара, кол-во купленного товара за год и годовая выручка) нулевая гипотеза была отвергнута. В основной группе количество заказов в 2.7 раза выше, чем в дополнительно группе. Средняя выручка для товара в основной группе больше более чем 7 раз. А самое главное, что средняя цена товаров из основной группы равна почти 1200 рублей и в четверо выше средней цены товаров из дополнительной группы (290 рублей).
