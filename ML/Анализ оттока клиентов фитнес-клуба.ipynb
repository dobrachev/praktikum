## Модель прогнозирования оттока пользователей<a id="step4"></a>
[*Назад к оглавлению*](#content)

Построим модель бинарной классификации пользователей, где целевой признак — факт оттока пользователя в следующем месяце.

Разобьём данные на обучающую и валидационную выборку(наблюдения независимы во времени):

```python
# разделим данные на признаки (матрица X) и целевую переменную (y)
X = gym_churn.drop(columns = ['churn'])
y = gym_churn['churn']

# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

Обучим модель на train-выборке двумя способами: логистической регрессией и случайным лесом.

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression()
# обучим модель
lr_model.fit(X_train, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test)

# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
# обучим модель случайного леса
rf_model.fit(X_train, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test)
```

Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке и сравним по ним модели.

```python
# определим функцию, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, title = 'Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.3f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.3f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.3f}'.format(recall_score(y_true, y_pred)))
```

```python
# выведем все метрики
print_all_metrics(y_test, lr_predictions, title='Метрики для модели логистической регрессии:')
print()
print_all_metrics(y_test, rf_predictions, title = 'Метрики для модели случайного леса:')
```

Посмотрим какие коэффициенты подобрала модель логистической регрессии для различных признаков:

```python
# создадим датафрейм с признаками и их весами
features = pd.DataFrame({'feature' : X.columns, 'coeff' : lr_model.coef_[0]})
features['coeff_abs'] = abs(features['coeff'])

# выведем упорядоченный по модулю коэффициентов датафрейм с признаками
features = features.sort_values(by = 'coeff_abs', ascending = False)
features
```

Построим модель логистической регрессии с использованием L1-регуляризации для определения наиболее значимых признаков, которые влияют на отток клиентов:

```python
# зададим алгоритм для модели логистической регрессии
lr_model_L1 = LogisticRegression(penalty = 'l1')
# обучим модель
lr_model_L1.fit(X_train, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions_L1 = lr_model_L1.predict(X_test)
```

```python
print_all_metrics(y_test, lr_predictions_L1, title='Метрики для модели логистической регрессии c L1-регуляризацией:')
```

```python
# создадим датафрейм с признаками и их весами
features_L1 = pd.DataFrame({'feature' : X.columns, 'coeff' : lr_model_L1.coef_[0]})
features_L1['coeff_abs'] = abs(features_L1['coeff'])


# выведем упорядоченный по модулю коэффициентов датафрейм с признаками
features_L1 = features_L1.sort_values(by = 'coeff_abs', ascending = False)
features_L1
```

## Кластеризация пользователей<a id="step5"></a>
[*Назад к оглавлению*](#content)

Сделаем кластеризацию пользователей, чтобы выделить несколько характерный групп, с которыми можно будет потом работать по отдельности.

Стандартизируем данные:

```python
# стандартизация данных перед работой с алгоритмом
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

Построим матрицу расстояний на стандартизованной матрице признаков и нарисуем дендрограмму:

```python
linked = linkage(X_sc, method = 'ward')

plt.figure(figsize=(15, 10))  
dendrogram(linked, color_threshold = 60, orientation='top')
plt.title('Иерархическая кластеризация пользователей')
plt.show()
```

Иерархическая кластеризация пользователей и дендрограмма говорят о том, что можно выделить примерно 4-6 групп. Если выделять четыре группы, то одна из групп будет слишком большой, поэтому хорошим решением будет выделить 5 групп пользователей.

Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. Иерархическая кластеризация говорит о том, что число кластеров можно принять равным 5:

```python
km = KMeans(n_clusters = 5, random_state = 0)
# прогнозируем кластеры для наблюдений
labels = km.fit_predict(X_sc)

# сохраняем метки кластера в поле нашего датасета
gym_churn['cluster_km'] = labels
```

Посмотрим на средние значения признаков для кластеров:

```python
gym_churn.groupby('cluster_km').mean().T
```

Отсортируем кластеры в порядке возростания оттока, пронумеровав их соответствующим образом:

```python
cluster_churn = pd.DataFrame(gym_churn.groupby('cluster_km')['churn'].mean()).sort_values(
    by = 'churn').reset_index().reset_index()
cluster_churn
```

```python
gym_churn['cluster'] = gym_churn['cluster_km'].apply(
    lambda x: cluster_churn.loc[cluster_churn['cluster_km'] == x, 'index'].to_numpy()[0])
```

```python
percent = gym_churn.groupby('cluster')['churn'].value_counts(normalize=True).mul(100).rename('percent').reset_index()
sns.catplot(x='cluster', y='percent', hue='churn', kind='bar', data=percent, height=5, aspect=2.5)
plt.title('Доли оттока для кластеров')
plt.xlabel('Кластер')
plt.ylabel('Доля, %')
plt.show()
```

```python
gym_churn.drop('cluster_km', axis = 1, inplace = True)
```

```python
gym_churn.groupby('cluster').mean().T
```

```python
gym_churn.groupby('cluster')['churn'].count()
```

Очевидные выводы:
 - кластер 0 и 1 - клиенты, для которых отток характерен с наименьшей вероятностью;
 - кластеры 3 и 4 - наибольшая вероятность оттока;
 - кластеры 2 - вероятность оттока такая же, какая для всего датасета.

При этом в 2 кластер попало меньше всего клиентов - и это люди без указанного номера телефона. Таким образом, алгоритм выделил особую группу, однако эта характерная особенность никак не влияет на отток.

Кластеры 0:
 - клиенты с самой большой длительностью контракта,
 - с высокой вероятностью пользующиеся партнерской программой или приглашениями друзей,
 - большие траты на дополнительные услуги фитнес-центра,
 - самая большая вероятность посещения групповых занятий,
 - посещаемость, которая не падает за последний месяц,
 - почти всегда живут или работают возле фитнес-центра.
 
Кластеры 1:
 - клиенты с самой высокой средней посещаемостью
 - средняя длительностью абонемента около 3 месяцев
 - значительно реже эти клиенты приходят по партнерской программе или приглашению друзей (они сами решили, что будут ходить)

Кластеры 2:
 - клиенты,которые не дали номер телефона
 - вероятность оттока такая же, какая для всего датасета
